```python
import matplotlib.pyplot as plt

def bresenham_circle(xc, yc, r):
    """
    Implements Bresenham's Circle Drawing Algorithm.

    Args:
        xc (int): X-coordinate of the circle's center.
        yc (int): Y-coordinate of the circle's center.
        r (int): Radius of the circle.

    Returns:
        list: A list of (x, y) tuples representing the points on the circle.
    """
    x = 0
    y = r
    d = 3 - 2 * r  # Initial decision parameter
    points = set()  # Use a set to avoid duplicate points

    def plot_symmetric_points(cx, cy, px, py):
        """Helper function to add all 8 symmetric points."""
        points.add((cx + px, cy + py))
        points.add((cx - px, cy + py))
        points.add((cx + px, cy - py))
        points.add((cx - px, cy - py))
        points.add((cx + py, cy + px))
        points.add((cx - py, cy + px))
        points.add((cx + py, cy - px))
        points.add((cx - py, cy - px))

    plot_symmetric_points(xc, yc, x, y)

    while y >= x:
        x += 1
        if d > 0:
            y -= 1
            d = d + 4 * (x - y) + 10
        else:
            d = d + 4 * x + 6
        plot_symmetric_points(xc, yc, x, y)

    return sorted(list(points)) # Return sorted list for consistent output

if __name__ == "__main__":
    center_x, center_y = 50, 50
    radius = 20

    circle_points = bresenham_circle(center_x, center_y, radius)

    # Separate x and y coordinates for plotting
    x_coords = [p[0] for p in circle_points]
    y_coords = [p[1] for p in circle_points]

    plt.figure(figsize=(9, 2))
    plt.scatter(x_coords, y_coords, s=10, color='blue')
    plt.title("Bresenham's Circle Drawing Algorithm")
    plt.xlabel("X-coordinate")
    plt.ylabel("Y-coordinate")
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='box') # Maintain aspect ratio for a true circle
    plt.show()
```
