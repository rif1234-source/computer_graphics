```python
import matplotlib.pyplot as plt

def dda_line_drawing(x1, y1, x2, y2):
    """
    Implements the DDA line drawing algorithm.

    Args:
        x1, y1: Coordinates of the starting point.
        x2, y2: Coordinates of the ending point.

    Returns:
        A tuple containing two lists: x_coords and y_coords,
        representing the pixels on the line.
    """
    dx = x2 - x1
    dy = y2 - y1

    # Determine the number of steps
    if abs(dx) > abs(dy):
        steps = abs(dx)
    else:
        steps = abs(dy)

    # Calculate increments for x and y
    x_increment = dx / steps
    y_increment = dy / steps

    x_coords = []
    y_coords = []

    x = x1
    y = y1

    # Plot the starting point
    x_coords.append(round(x))
    y_coords.append(round(y))

    # Iterate and plot subsequent points
    for _ in range(int(steps)):
        x += x_increment
        y += y_increment
        x_coords.append(round(x))
        y_coords.append(round(y))

    return x_coords, y_coords

if __name__ == "__main__":
    # Example usage
    x1, y1 = 2, 5
    x2, y2 = 4, 8

    x_pixels, y_pixels = dda_line_drawing(x1, y1, x2, y2)

    print(f"X-coordinates: {x_pixels}")
    print(f"Y-coordinates: {y_pixels}")

    # Plotting the line using Matplotlib
    plt.plot(x_pixels, y_pixels, 'o-', markersize=4) # 'o-' for points connected by line
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("DDA Line Drawing Algorithm")
    plt.grid(True)
    plt.show()
```
