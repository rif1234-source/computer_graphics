```python
import numpy as np
import matplotlib.pyplot as plt

def reflect_2d(points, reflection_matrix):
    """Applies a 2D reflection transformation to a set of points."""
    return np.dot(points, reflection_matrix)

# Define an example object (e.g., a triangle)
original_points = np.array([
    [1, 1],
    [3, 1],
    [2, 3]
])

# Define reflection matrices
reflection_x_axis = np.array([[1, 0], [0, -1]])
reflection_y_axis = np.array([[-1, 0], [0, 1]])
reflection_origin = np.array([[-1, 0], [0, -1]])

# Perform reflections
reflected_x = reflect_2d(original_points, reflection_x_axis)
reflected_y = reflect_2d(original_points, reflection_y_axis)
reflected_origin = reflect_2d(original_points, reflection_origin)

# Plotting
plt.figure(figsize=(10, 5))

# Original object
plt.subplot(1, 3, 1)
plt.plot(original_points[:, 0], original_points[:, 1], 'o-')
plt.title('Original Object')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# Reflection about X-axis
plt.subplot(1, 3, 2)
plt.plot(original_points[:, 0], original_points[:, 1], 'o-', label='Original')
plt.plot(reflected_x[:, 0], reflected_x[:, 1], 'o-', label='Reflected (X-axis)')
plt.title('Reflection about X-axis')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# Reflection about Y-axis
plt.subplot(1, 3, 3)
plt.plot(original_points[:, 0], original_points[:, 1], 'o-', label='Original')
plt.plot(reflected_y[:, 0], reflected_y[:, 1], 'o-', label='Reflected (Y-axis)')
plt.title('Reflection about Y-axis')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.xlim(-5, 5)
plt.ylim(-5, 5)

plt.tight_layout()
plt.show()
```
