```python
import numpy as np
import matplotlib.pyplot as plt

def rotate_point(point, angle_degrees, pivot_point=(0, 0)):
    """
    Rotates a 2D point around a specified pivot point.

    Args:
        point (tuple): The (x, y) coordinates of the point to rotate.
        angle_degrees (float): The rotation angle in degrees (counter-clockwise).
        pivot_point (tuple): The (x, y) coordinates of the pivot point.

    Returns:
        tuple: The (x', y') coordinates of the rotated point.
    """
    x, y = point
    xp, yp = pivot_point
    angle_radians = np.radians(angle_degrees)

    # Translate to origin
    x_shifted = x - xp
    y_shifted = y - yp

    # Rotate
    x_rotated = x_shifted * np.cos(angle_radians) - y_shifted * np.sin(angle_radians)
    y_rotated = x_shifted * np.sin(angle_radians) + y_shifted * np.cos(angle_radians)

    # Translate back
    x_prime = x_rotated + xp
    y_prime = y_rotated + yp

    return x_prime, y_prime

# Example Usage: Rotate a square
if __name__ == "__main__":
    # Define the vertices of a square
    square_vertices = np.array([[1, 1], [3, 1], [3, 3], [1, 3], [1, 1]]) # Close the loop

    # Define rotation parameters
    rotation_angle = 45  # degrees
    pivot = (2, 2)       # Center of the square

    # Rotate each vertex
    rotated_square_vertices = np.array([rotate_point(vertex, rotation_angle, pivot) for vertex in square_vertices])

    # Plotting
    plt.figure(figsize=(6, 6))
    plt.plot(square_vertices[:, 0], square_vertices[:, 1], 'b-', label='Original Square')
    plt.plot(rotated_square_vertices[:, 0], rotated_square_vertices[:, 1], 'r-', label='Rotated Square')
    plt.plot(pivot[0], pivot[1], 'go', markersize=8, label='Pivot Point')

    plt.title("2D Rotation Transformation")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid(True)
    plt.axhline(0, color='gray', linewidth=0.5)
    plt.axvline(0, color='gray', linewidth=0.5)
    plt.legend()
    plt.axis('equal') # Ensure equal scaling on both axes
    plt.show()
```
