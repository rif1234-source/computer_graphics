```python
import matplotlib.pyplot as plt

def shear_x(points, shx):
    """Applies X-shearing to a list of 2D points."""
    sheared_points = []
    for x, y in points:
        new_x = x + shx * y
        sheared_points.append((new_x, y))
    return sheared_points

def shear_y(points, shy):
    """Applies Y-shearing to a list of 2D points."""
    sheared_points = []
    for x, y in points:
        new_y = y + shy * x
        sheared_points.append((x, new_y))
    return sheared_points

def draw_polygon(points, color='blue', label=''):
    """Draws a polygon from a list of points."""
    x_coords = [p[0] for p in points]
    y_coords = [p[1] for p in points]
    plt.plot(x_coords + [x_coords[0]], y_coords + [y_coords[0]], color=color, label=label) # Close the polygon

# Example Usage:
if __name__ == "__main__":
    # Define a sample square
    square_points = [(1, 1), (3, 1), (3, 3), (1, 3)]

    plt.figure(figsize=(8, 6))
    plt.grid(True)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("2D Shearing Transformation")
    plt.axis('equal') # Ensure equal scaling on both axes

    # Draw original square
    draw_polygon(square_points, color='blue', label='Original Square')

    # Apply X-shearing
    shx_factor = 0.5
    sheared_x_square = shear_x(square_points, shx_factor)
    draw_polygon(sheared_x_square, color='red', label=f'X-Sheared (shx={shx_factor})')

    # Apply Y-shearing (on original square for clarity)
    shy_factor = 0.5
    sheared_y_square = shear_y(square_points, shy_factor)
    draw_polygon(sheared_y_square, color='green', label=f'Y-Sheared (shy={shy_factor})')

    plt.legend()
    plt.show()
```
